```python
import pandas as pd
import numpy as np

#df = pd.read_csv('/content/diversity_text.csv', encoding='cp1252')
df = pd.read_csv('/content/diversity_text.csv')
```

```python
#sentences = []
hetero_text_summary =[]
for i in range(len(df)):

  sentences = []
  sentences.append(df.iloc[:,1][i])
  x = summarizer(sentences)
  hetero_text_summary.append(x)

#
```

```python
homo_text_summary =[]
for i in range(len(df)):
  sentences = []
  sentences.append(df.iloc[:,2][i])
  y = summarizer(sentences)
  homo_text_summary.append(y)
```

```python
hetero_text_summary1 = []
for i in range(len(hetero_text_summary)):
    hetero_text_summary1.append(hetero_text_summary[i][0]['summary_text'])
```

```python
homo_text_summary1 = []
for i in range(len(homo_text_summary)):
    homo_text_summary1.append(homo_text_summary[i][0]['summary_text'])
```

```python
print(homo_text_summary1)
```

```python
df1 =  pd.DataFrame(hetero_text_summary1,columns =['hetero_text_summary'], index = None)
df2 =  pd.DataFrame(homo_text_summary1,columns =['homo_text_summary'], index = None)
```

```python
dataframe = df.copy(deep=True)
```

```python
dataframe.head(1)
```

```python
len(dataframe)
```

```python
dataframe['hetero_text_summary'] = df1['hetero_text_summary']

dataframe['homo_text_summary'] = df2['homo_text_summary']
```

```python
dataframe.head(1)
```

```python
!pip install sentence_transformers
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
model = SentenceTransformer('sentence-transformers/paraphrase-MiniLM-L6-v2')
```

```python
dataframe.head(1)
```

```python
hetero_BART_sim1 = []

for i in range(len(dataframe)):
    sentences = []
    sentences.append(dataframe.iloc[:,1][i])
    sentences.append(dataframe.iloc[:,3][i])
    embeddings = model.encode(sentences)
    score = cosine_similarity(embeddings[[0]],embeddings[[1]])
    s = score.tolist() 
    hetero_BART_sim1.append(s[0][0]) 
```

```python
homo_BART_sim1 = []

for i in range(len(dataframe)):
    sentences = []
    sentences.append(dataframe.iloc[:,2][i])
    sentences.append(dataframe.iloc[:,4][i])
    embeddings = model.encode(sentences)
    score = cosine_similarity(embeddings[[0]],embeddings[[1]])
    s = score.tolist() 
    homo_BART_sim1.append(s[0][0]) 
```

```python
model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
```

```python
hetero_BART_sim2 = []

for i in range(len(dataframe)):
    sentences = []
    sentences.append(dataframe.iloc[:,1][i])
    sentences.append(dataframe.iloc[:,3][i])
    embeddings = model.encode(sentences)
    score = cosine_similarity(embeddings[[0]],embeddings[[1]])
    s = score.tolist() 
    hetero_BART_sim2.append(s[0][0]) 
```

```python
homo_BART_sim2 = []

for i in range(len(dataframe)):
    sentences = []
    sentences.append(dataframe.iloc[:,2][i])
    sentences.append(dataframe.iloc[:,4][i])
    embeddings = model.encode(sentences)
    score = cosine_similarity(embeddings[[0]],embeddings[[1]])
    s = score.tolist() 
    homo_BART_sim2.append(s[0][0]) 
```

```python
output = pd.DataFrame(list(zip(hetero_BART_sim1, homo_BART_sim1, hetero_BART_sim2, homo_BART_sim2)),
               columns =['hetero_BART_sim1', 'homo_BART_sim1', 'hetero_BART_sim2', 'homo_BART_sim2'])
```
