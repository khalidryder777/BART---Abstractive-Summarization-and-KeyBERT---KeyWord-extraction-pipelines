## Removing 'Amani','Alex','Max','Sam','Jordan','Taylor','Jessie','Jude','Dion','Ariel','Lian','Avery'



```python
df.head(1)
```

```python
import re

names_to_remove = ['Amani P.', 'Alex V.', 'Max M.', 'Sam V.', 'Jordan T.', 'Taylor N.', 'Jessie A.', 'Jude L.', 'Dion L.', 'Ariel C.', 'Lian C.', 'Avery D.',
                   'Amani','Alex','Max','Sam','Jordan','Taylor','Jessie','Jude','Dion','Ariel','Lian','Avery']

for name in names_to_remove:
    df["hetero_text_KeyBERT_Preprocessed"] = df["hetero_text"].apply(lambda x: re.sub(r"\b"+name+r"(?=\b)", "", x, 
                                                   flags=re.IGNORECASE))

for name in names_to_remove:
    df["homo_text_KeyBERT_Preprocessed"] = df["homo_text"].apply(lambda x: re.sub(r"\b"+name+r"(?=\b)", "", x, 
                                                   flags=re.IGNORECASE))

```

```python
df.head(1)
```

### Preprocessing the text 


1.   Tokenization
2.   Lower Casing
3.   Removing Stop Words
4.   Lemmatization



#### Removing punctuation as well

```python
import nltk
import string
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')

stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()





df["hetero_text_KeyBERT_Preprocessed"] = df["hetero_text_KeyBERT_Preprocessed"].apply(lambda x: 
                                            [lemmatizer.lemmatize(word.lower()) 
                                            for word in word_tokenize(x) 
                                            if (word.lower() not in stop_words)
                                            and (re.match("^[a-zA-Z]+$", word))
                                            ]) 
                                            

df["homo_text_KeyBERT_Preprocessed"] = df["homo_text_KeyBERT_Preprocessed"].apply(lambda x: 
                                            [lemmatizer.lemmatize(word.lower()) 
                                            for word in word_tokenize(x) 
                                            if (word.lower() not in stop_words)
                                            and (re.match("^[a-zA-Z]+$", word))
                                            ])

```

```python
!pip install sentence_transformers
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
model = SentenceTransformer('sentence-transformers/paraphrase-MiniLM-L6-v2')
```

```python
dataframe.head(1)
```

```python
hetero_KeyBERT_sim1 = []

for i in range(len(dataframe)):
    sentences = []
    sentences.append(dataframe.iloc[:,1][i])
    sentences.append(dataframe.iloc[:,5][i])
    embeddings = model.encode(sentences)
    score = cosine_similarity(embeddings[[0]],embeddings[[1]])
    s = score.tolist() 
    hetero_KeyBERT_sim1.append(s[0][0]) 
```
